# Dólar 

```python
import yfinance as yf
import matplotlib.pyplot as plt
import mplcursors

# Baixar os dados do USD/BRL
usbrl = yf.download('USDBRL=X', start='2000-01-01')  # Ajuste a data de início conforme necessário

# Plotar os dados
plt.figure(figsize=(12, 6))
plt.plot(usbrl['Close'], label='USDBRL=X - Fechamento', color='blue')  # Linha azul
plt.title('Cotação USD/BRL')
plt.xlabel('Data')
plt.ylabel('Preço (R$)')
plt.legend()
plt.grid(True)

# Habilitar funcionalidade de hover com mplcursors
cursor = mplcursors.cursor(hover=True)
cursor.connect(
    "add", lambda sel: sel.annotation.set_text(f"Preço: R${sel.ydata:.2f}\nData: {sel.target[0].strftime('%Y-%m-%d')}")
)

plt.show()
```

```python
import pandas_datareader.data as web
import datetime

# Definir a data de início
start = datetime.datetime(2000, 1, 1)

# Baixar os dados do Payroll
Dolar = web.DataReader('dexbzus', 'fred', start)

# Exibir os dados
print(Dolar)
```
